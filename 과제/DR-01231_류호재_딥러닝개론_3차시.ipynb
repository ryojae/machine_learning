# 최소 제곱법

```python
import numpy as np

x= np.array([2, 4, 6, 8])
y= np.array([81, 93, 91, 97])

mx= np.mean(x)
my= np.mean(y)

divisor = sum([(i - mx)**2 for i in x])

def top(x, mx, y, my):
    d= 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my)
    return d
dividend= top(x, mx, y, my)

a = dividend / divisor
b = my - (mx*a)

print(f'기울기: {a}')
print(f'절편: {b}')
```

# 평균 제곱 오차

```python
import numpy as np

fake_a = 3
fake_b = 76

def predict(x):
    return fake_a * x + fake_b

x= np.array([2, 4, 6, 8])
y= np.array([81, 93, 91, 97])

predict_result = []

for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print(f'공부시간={x[i]}, 실제 점수={y[i]}, 예측 점수={predict(x[i])}')

n= len(x)
def mse(y, y_pred):
    return (1/n) * sum((y - y_pred)**2)

print(f'평균 제곱 오차: {mse(y, predict_result)}')
```

# 경사하강법

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2,4,6,8])
y = np.array([81, 93, 91, 97])

plt.scatter(x,y)
plt.show()

a = 0
b = 0
lr = 0.03

epochs = 2001

n = len(x)

for i in range(epochs):
    y_pred = a * x + b
    error = y - y_pred

    a_diff = (2/n) * sum(-x * error)
    b_diff = (2/n) * sum(-(error))

    a = a - lr * a_diff
    b = b - lr * b_diff

    if i % 100 == 0:
        print(f'epoch={i}, 기울기={a:.4f}, 절편={b:.4f}')
    # plt.plot(x,y_pred,'r')
y_pred = a * x + b

plt.scatter(x, y, c='blue')
plt.plot(x,y_pred,'r')
plt.show()
```

#다중선형회귀

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
boston = fetch_california_housing()
print(boston.DESCR)
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['PRICE'] = boston.target
boston_df.head()
```

```python
print('보스톤 주택 가격 데이터셋 크기: ',boston_df.shape)
boston_df.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

Y = boston_df['PRICE']
X = boston_df.drop(['PRICE'],axis=1,inplace=False)

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,
                                    test_size=0.3,random_state=156)

lr = LinearRegression()
lr.fit(X_train,Y_train)

Y_predict = lr.predict(X_test)

mse = mean_squared_error(Y_test,Y_predict)
rmse = np.sqrt(mse)

print(f'MSE: {mse:.3f}, RMSE: {rmse:.3f}')
print('R^2: {:.3f}'.format(r2_score(Y_test,Y_predict)))
```

```python
print('Y 절편 값:',lr.intercept_)
print('회귀 계수 값:',np.round(lr.coef_,1))
```

```python
coef = pd.Series(data=np.round(lr.coef_,2),index=X.columns)
coef.sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(figsize = (8, 8), ncols = 2, nrows = 4)
x_features = ['AveBedrms', 'MedInc', 'HouseAge', 'Population', 'AveOccup', 'AveRooms', 'Latitude', 'Longitude']
for i, feature in enumerate(x_features):
  row = int(i/2)
  col = i%2
  sns.regplot(x = feature, y = 'PRICE', data = boston_df, ax = axs[row][col])
```

# 다층 퍼셉트론

```python
import numpy as np

w11 = np.array([-2,-2])
w12 = np.array([2,2])
w2 = np.array([1,1])
b1 = 3
b2 = -1
b3 = -1

def MLP(x,w,b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1

def NAND(x1,x2):
    return MLP(np.array([x1,x2]),w11,b1)

def OR(x1,x2):
    return MLP(np.array([x1,x2]),w12,b2)

def AND(x1,x2):
    return MLP(np.array([x1,x2]),w2,b3)

def XOR(x1,x2):
    return AND(NAND(x1,x2),OR(x1,x2))

for x in [(0,0),(1,0),(0,1),(1,1)]:
    y = XOR(x[0],x[1])
    print(f'입력값: {x}, 출력값: {y}')
```

# 자동차 연비 계산

```python
import numpy as np
import pandas as pd

data_df = pd.read_csv('/content/drive/MyDrive/두산_로보틱스_머신러닝/auto-mpg.csv',header= 0, engine='python')
print('데이터셋 크기:',data_df.shape)
data_df.head()
```

```python
data_df = data_df.drop(['car_name','origin','horsepower'], axis=1,inplace = False)
data_df.head()
```

```python
print(data_df.shape)
data_df.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

Y = data_df['mpg']
X = data_df.drop(['mpg'],axis=1,inplace=False)

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)

lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
Y_predict = lr.predict(X_test)

mse = mean_squared_error(Y_test,Y_predict)
rmse = np.sqrt(mse)

print(f'MSE: {mse:.3f}, RMSE: {rmse:.3f}')
print('R^2: {:.3f}'.format(r2_score(Y_test,Y_predict)))
```

```python
print('Y 절편 값:',np.round(lr.intercept_,2))
print('회귀 계수 값:',np.round(lr.coef_,2))
```

```python
coef = pd.Series(data = np.round(lr.coef_,2),index = X.columns)
coef.sort_values(ascending=False)
```
